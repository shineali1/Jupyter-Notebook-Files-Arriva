```python
import pandas as pd

# Read in the datasets
df1 = pd.read_excel("F:\Career\Thesis Projects\Jupyter Notebook Files Arriva\Fuel_consumption_report.xlsx")
df2 = pd.read_excel('F:\Career\Thesis Projects\Jupyter Notebook Files Arriva\Trip_report.xlsx')
df3 = pd.read_excel('F:\Career\Thesis Projects\Jupyter Notebook Files Arriva\Idle_fuel_exception_report.xlsx')

# Convert date and time columns to datetime objects
df1['Date'] = pd.to_datetime(df1['Date'], format='%d-%m-%Y')
df2['Day'] = pd.to_datetime(df2['Day'], format='%a %d/%m')
df2['Start time'] = pd.to_datetime(df2['Start time'], format='%H:%M:%S')
df2['End time'] = pd.to_datetime(df2['End time'], format='%H:%M:%S')
df3['Date'] = pd.to_datetime(df3['Date'], format='%a %d-%m')
df3['Start time'] = pd.to_datetime(df3['Start time'], format='%H:%M:%S')
df3['End time'] = pd.to_datetime(df3['End time'], format='%H:%M:%S')

# Create three dataframes
df_fuel_consumption = df1[['Vehicle', 'Date', 'Distance [km]', 'Duration [h]', 'Standstill [h]', 'Fuel consumption [l]', 'Average fuel consumption [l/100km]', '+/- Reference (absolute) [l/100km]', '+/- Reference (relative) [%]', 'CO2 emission [kg]']]
df_driving_data = df2[['Vehicle', 'Day', 'Start time', 'End time', 'Start odometer [km]', 'Driver', 'Start location', 'Fuel consumption [l]', 'Average fuel consumption [l/100km]', '+/- Reference (absolute) [l/100km]', '+/- Reference (relative) [%]']]
df_idle_data = df3[['Vehicle', 'Date', 'Start time', 'End time', 'Idle time [h]', 'Wasted fuel [l]', 'CO2 emission [kg]', 'Location']]

```

```python
#Data cleaning for Fuel Consumption Dataframe
# remove any rows with missing values
df_fuel_consumption.dropna(inplace=True)

# remove any rows with zero distance or duration
df_fuel_consumption = df_fuel_consumption[(df_fuel_consumption['Distance [km]'] != 0) & (df_fuel_consumption['Duration [h]'] != '00:00:00')]

# convert Duration and standstill columns to timedelta objects
df_fuel_consumption['Duration [h]'] = df_fuel_consumption['Duration [h]'].astype(str)
df_fuel_consumption['Standstill [h]'] = df_fuel_consumption['Standstill [h]'].astype(str)
df_fuel_consumption['Duration [h]'] = df_fuel_consumption['Duration [h]'].str.replace('-', '') # remove negative signs
df_fuel_consumption['Standstill [h]'] = df_fuel_consumption['Standstill [h]'].str.replace('-', '') # remove negative signs

df_fuel_consumption['Duration [h]'] = pd.to_timedelta(df_fuel_consumption['Duration [h]'])
df_fuel_consumption['Standstill [h]'] = pd.to_timedelta(df_fuel_consumption['Standstill [h]'])

# convert Fuel consumption column to float
df_fuel_consumption['Fuel consumption [l]'] = df_fuel_consumption['Fuel consumption [l]'].astype(float)

# convert CO2 emission column to float
df_fuel_consumption['CO2 emission [kg]'] = df_fuel_consumption['CO2 emission [kg]'].astype(float)

# drop any unnecessary columns
df_fuel_consumption = df_fuel_consumption.drop(['+/- Reference (absolute) [l/100km]', '+/- Reference (relative) [%]'], axis=1)

# rename columns to remove spaces and special characters
df_fuel_consumption = df_fuel_consumption.rename(columns={'Vehicle': 'vehicle', 'Date': 'date', 'Distance [km]': 'distance', 'Duration [h]': 'duration', 
                        'Standstill [h]': 'standstill', 'Fuel consumption [l]': 'fuel_consumption', 
                        'Average fuel consumption [l/100km]': 'average_fuel_consumption', 
                        'CO2 emission [kg]': 'co2_emission'})

# reset the index
df_fuel_consumption = df_fuel_consumption.reset_index(drop=True)

df_fuel_consumption.head()
```

```python
#Data Cleaning for Driving Data Dataframe
# drop any irrelevant columns
df_driving_data = df_driving_data.drop(columns=['Vehicle', 'Driver', 'Start location', '+/- Reference (absolute) [l/100km]', '+/- Reference (relative) [%]'])

# rename the columns for easier access
df_driving_data = df_driving_data.rename(columns={'Day': 'date', 'Start time': 'start_time', 'End time': 'end_time', 'Start odometer [km]': 'start_odometer', 'Fuel consumption [l]': 'fuel_consumption', 'Average fuel consumption [l/100km]': 'average_fuel_consumption'})

# convert date, start_time, and end_time columns to datetime type
df_driving_data['date'] = pd.to_datetime(df_driving_data['date'])
df_driving_data['start_time'] = pd.to_datetime(df_driving_data['start_time'], format='%Y-%m-%d %H:%M:%S')
df_driving_data['end_time'] = pd.to_datetime(df_driving_data['end_time'], format='%Y-%m-%d %H:%M:%S')

# remove any rows with missing values
df_driving_data = df_driving_data.dropna()

# remove any rows with zero or negative fuel consumption values
df_driving_data = df_driving_data[df_driving_data['fuel_consumption'] > 0]

# calculate the distance traveled based on the odometer readings
df_driving_data['distance'] = df_driving_data['start_odometer'].diff().fillna(0)

# calculate the actual fuel consumption in liters per 100 kilometers
df_driving_data['actual_fuel_consumption'] = 100 * df_driving_data['fuel_consumption'] / df_driving_data['distance']

df_anomaly = df_driving_data[df_driving_data['actual_fuel_consumption'] >= 100]
# remove any rows with actual fuel consumption greater than 100 liters per 100 kilometers (anomaly)
df_driving_data = df_driving_data[df_driving_data['actual_fuel_consumption'] <= 100]

# reset the index
df_driving_data = df_driving_data.reset_index(drop=True)
```
